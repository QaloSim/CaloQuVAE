# testing the metric saving:

```python
# input the path with the Frobenius metrics
# running the script will make this file:
path = "/fast_scratch_1/caloqvae/dkraft/wandb/run-20250718_204309-ylzyaumb/files/Frobenius_All.npz"

def describe_npz(path, preview=100):
    d = np.load(path, allow_pickle=True) # load
    print(f"\nFile: {os.path.basename(path)}") # filename
    print("Keys:", ", ".join(sorted(d.files))) # print ALL keys

    # epochs summary if many epochs are saved
    if "epochs" in d:
        e = d["epochs"]
        print(f"\nepochs: shape={e.shape}, dtype={e.dtype},"
              f"min={e.min() if e.size else 'NA'}, max={e.max() if e.size else 'NA'}")

    print("\n Per-key details")
    for k in sorted(d.files):
        arr = d[k]
        if isinstance(arr, np.ndarray):
            print(f"- {k}: shape={arr.shape}, dtype={arr.dtype}")
            if arr.dtype == object:
                # show inner shapes too
                inner = []
                for x in arr[:preview]:
                    if isinstance(x, np.ndarray):
                        inner.append(f"ndarray{list(x.shape)} dtype={x.dtype}")
                    else:
                        inner.append(type(x).__name__)
                print(f" Sample elements (first {min(preview, len(arr))}): {inner}")
            else:
                # some stats (min/max)
                flat = arr.ravel()
                if flat.size:
                    print(f"  min={np.nanmin(flat):.6g}, max={np.nanmax(flat):.6g}, mean={np.nanmean(flat):.6g}")

    return d # to check specific variables

# uncomment to check output:
#d = describe_npz(path)
```
